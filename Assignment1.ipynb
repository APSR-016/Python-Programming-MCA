```python
# taking first name as an input
first_str=input('Enter first name:')
# taking last name as an input
last_str=input('Enter last name:')
first_rev="".join(reversed(first_str))# reversed function can be used to reverse the string
last_rev="".join(reversed(last_str))# join function can be used to join characters or strings to one string
full_name=" ".join([first_rev,last_rev])
print('Full name in reverse is:',full_name)
```

```python
# Taking input from the user
user_input = input("Enter a number: ")

# Converting the input to different types
try:
    int_value = int(user_input)
    print("Integer value:", int_value)
except ValueError:
    print("Cannot convert to an integer.")

try:
    float_value = float(user_input)
    print("Float value:", float_value)
except ValueError:
    print("Cannot convert to a float.")

try:
    complex_value = complex(user_input)
    print("Complex value:", complex_value)
except ValueError:
    print("Cannot convert to a complex number.")
```

```python
# an endless loop that only exits if the flag is entered as 'exit'
while(True):
    num=input('Enter a number:')
    if num[0]=='-':     # checks for negative number
        print('The entered number is NEGATIVE.')
    elif num[0]=='0':   # checks if the number is zero
        print('The entered number is ZERO.')
    else:               # checks if the number is positive
        print('The entered number is POSITIVE.')
    flag=input(print('To exit the program enter- exit'))
    if flag.lower()=='exit':
        break   # terminates the loop
    else:
        continue   # continues the iteration of the loop
```

```python
# function to input two numbers and check if they are odd or even
def oddeve(num1,num2):
    if num1%2==0 and num2%2==0:
        print('Both numbes are Even')
    elif num1%2!=0 and num2%2!=0:
        print('Both numbers are Odd')
    elif num1%2==0 and num2%2!=0:
        print('First number- Even\nSecond number- Odd')
    else:
        print('First number- Odd\nSecond number- Even')

# taking both the numbers as input
num1=int(input('Enter first number:'))
num2=int(input('Enter second number:'))
oddeve(num1,num2)
```

```python
# Taking an integer input from the user
num = int(input("Enter an integer: "))

# Function to convert to binary using bitwise operators
def int_to_binary(n):
    binary = ""
    for i in range(n.bit_length() - 1, -1, -1):
        binary += "1" if (n & (1 << i)) else "0"
    return binary or "0"

# Function to convert to octal using bitwise operators
def int_to_octal(n):
    octal = ""
    temp = n
    while temp > 0:
        octal = str(temp & 0o7) + octal
        temp >>= 3
    return octal or "0"

# Function to convert to hexadecimal using bitwise operators
def int_to_hexadecimal(n):
    hex_digits = "0123456789ABCDEF"
    hexadecimal = ""
    temp = n
    while temp > 0:
        hexadecimal = hex_digits[temp & 0xF] + hexadecimal
        temp >>= 4
    return hexadecimal or "0"

# Displaying the results
print("Binary equivalent:", int_to_binary(num))
print("Octal equivalent:", int_to_octal(num))
print("Hexadecimal equivalent:", int_to_hexadecimal(num))
```
